## Histogram

```python
sns.histplot(data=data,x='SepalLengthCm')
```

## Probability Density Function

Gráfico para função de densidade de probabilidade: 

```python
sns.kdeplot(data=data,x='PetalLengthCm')
```

```python
# dividindo por espécies
sns.kdeplot(data=data,x='PetalLengthCm',hue='Species')
```

## Boxplot
Boxplot permite vizualizar a distribuição dos dados, verificar seus extremos e outliers

```python
sns.boxplot(data=data,y='PetalLengthCm',x='Species')
```

## Violinplot

Plot violino é similar ao boxplot com a adição da curva de densidade

```python
sns.violinplot(data=data,y='PetalLengthCm',x='Species')
```

## Pairplot
Função para plotar vários gráficos que relaciona as combinações das variáveis. Gerando em sua diagonal principal o gráfico de histograma e gráficos de disperções fora da diagonal principal. 

```python
sns.pairplot(data=df)
```

```python
# separando por espécies
sns.pairplot(data=data,hue='Species')
```

## Heatmap

Matrix de coeficientes de correlação Pearson plotados num gráfico de calor: 

```python
sns.heatmap(data=df.drop(columns='Species').corr(), annot=True,cmap = "GnBu")
```

## Joinplot
Joinplot plota a dispersão entre duas variáveis x e y e também seus respectivos histogramas ou densidade de probabilidade. 

```python
sns.jointplot(data=df,x='PetalLengthCm',y='SepalWidthCm')
```

```python
sns.jointplot(data=data,x='PetalLengthCm',y='SepalWidthCm',hue='Species')
```

## Rugplot

É similar ao joinplot, esse gráfico mostra a distribuição marginal das variáveis x e y. 

```python
sns.rugplot(data=data,x='PetalLengthCm',hue='Species')
```

```python
sns.rugplot(data=data,x='PetalLengthCm',y='SepalLengthCm',hue='Species')
```
